# Digital Advertising Spend & Campaign Revenue: A Regression Analysis

## Author
Andrej Arsenijevic, Dubai UAE, April 2026

## Problem Statement
Does an increase in digital ad spend across platforms, industries, and global 
markets drive measurable revenue growth for brands?

## Context
Advertising investment decisions impact billions in brand budgets worldwide.
Understanding which digital channels, campaign types, and markets deliver the 
strongest revenue returns enables a smarter, evidence-backed marketing strategy.
This analysis uses real campaign-level data spanning 7 countries and 3 major 
platforms to quantify the relationship between spend inputs and revenue outcomes.

## Target Variable (Y)
**Campaign Revenue (USD)** — a continuous numerical outcome ranging from $142.69 
to $295,028, suitable for regression modeling.

## Features / Independent Variables (X)
- Ad spend (USD) — primary investment input
- Clicks — direct engagement signal
- Impressions — reach and exposure signal
- Conversions — campaign response count
- CTR (click-through rate) — spend efficiency signal
- CPA (cost per acquisition) — cost efficiency control
- Date — time control for seasonality (full year 2024)
- Platform (categorical: Google Ads, Meta Ads, TikTok Ads)
- Campaign type (categorical: Search, Display, Video, Shopping)
- Industry (categorical: Fintech, SaaS, Healthcare, EdTech, E-commerce)
- Country (categorical control: USA, UK, UAE, Germany, Canada, India, Australia)

## Hypothesis
**H₀:** Ad spend, clicks, and conversions have no statistically significant 
effect on campaign revenue across platforms, industries, and markets.

**H₁:** Conversions and clicks are the strongest positive predictors of revenue, 
with ad spend showing a significant but smaller direct effect, and platform type 
and industry moderating the relationship — after controlling for country, 
campaign type, and seasonality.

## Dataset
- **Source:** Global Ads Performance Dataset — Kaggle
- **License:** Apache 2.0 — permitted for academic and student use
- **Scope:** 7 countries · 3 platforms · 5 industries · 4 campaign types
- **Period:** January 2024 — December 2024
- **Rows / Features:** 1,800 rows · 14 columns · 0 missing values

## Stakeholders
Marketing teams, brand strategists, and academic researchers studying the 
revenue impact of digital advertising spend across global markets and platforms.
